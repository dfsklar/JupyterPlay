# Import Station Info

```python
from os import listdir
from os.path import isfile, join
import numpy as np
import pandas as pd
import json
```

```python
with open("citibikedata/station_information.json", "r") as read_file:
        data = json.load(read_file)
df = pd.DataFrame(data['data']['stations'])
# So this is interesing, by default when importing an arrow, pandas will simply
# have this DF be indexed by sequence ID, rather meaningless.

# We want station_id to be the index, but first, we must make station_id to
# an integer to facilitate joins from external tables.
df = df.astype({'station_id': 'int32'})
```

```python
df = df.set_index('station_id')
df.dtypes
```
